# WeRateDogs Tweet Analysis
In this project, I analyzed the @WeRateDogs twitter account's tweets using the following sources:
- Archived Twitter Data (provided as a "file on hand" by Udacity)
- Image Predictions for Tweets (requested from a Udacity-hosted neural network)
- Additional Tweet Metadata (queried via Twitter API)

```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading in the dataset for graphs
df = pd.read_csv('tweet_data.csv', parse_dates=['created_at'])
```

In preparing this data, I gathered the data from the above sources, and then performed a variety of manipulations to ensure the data was both clean and tidy. This allows us to examine some key points of interest in the data, such as relatoinships between key metrics of a tweets popularity such as **favorite count** and **retweet_count** compared with things like the time the tweet was created and the rating given to the featured dog. Also, I ran a quick analysis on the most common dogs featured in WeRateDogs tweets, as predicted by a Udacity neural network.

## These are, probably, the most popular dogs to be featured in images of WeRateDogs tweets:
1. Golden Retriever
2. Labrador Retriever
3. Pembroke
4. Chihuahua
5. Pug

## Using an OLS linear regression model, I found that retweet count is a good predictor of favorite count (r-squared = .858), with each additional retweet predicting ~2.5 more favorites. Compared to retweet count, the time of the tweet and the dog rating were not correlated to favorite count well at all (r-squared < .5).

## Below, I show a graph of the number of favorites a tweet received by the rating given to the dog in the tweet (out of 10).

```python
df.plot.scatter('rating_numerator','favorite_count', figsize=(8,8));
plt.xlabel('Rating Given in Tweet (/10)');
plt.ylabel('# of Favorites');
plt.title('WeRateDog Favorites on Tweet Rating');
```

## At first, this seems a clear-cut case that the higher tha rating given in the tweet, the more popular a tweet is (with # of Favorites as the gauge of popularity). However, the curve also looks somewhat exponential, which seemed strange to me. So, I plotted the dog ratings over time below.

```python
f, ax = plt.subplots(figsize=(8, 8))
plt.scatter(df.created_at, df.rating_numerator)
plt.xticks(rotation=60)
plt.xlabel('Tweet Timestamp')
plt.ylabel('Rating Given in Tweet (/10)')
plt.title('WeRateDog Ratings Over Time');
```

## The graph above clearly shows a marked inflation of dog ratings over the span of 2015 to 2017. This supports the idea that any correlation between dog ratings and popularity of the tweet is coincidental. I hypothesize that the increase we see is actually due to exponential growth in popularity of the @WeRateDogs twitter account over time. Since the dog ratings also inflate over time, this leads to it seeming as though higher ratings yield more popular tweets. However, my investigation does not support this.
